## Collaborative Filtering Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(x[x].index)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
CF = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
CF.fillna(0,inplace=True)
```

```python
CF
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(CF)
```

```python
similarity_scores
```

```python
similarity_scores.shape
```

## Calculate shortest distance between books to recommend similar books, shorter the distance similar the book

```python
def recommend(book_name):
    # index fetch
    index = np.where(CF.index==book_name)[0][0]
    #to calculate shortest distance between books to recommend similar books, shorter the distance similar the book
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5] 
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == CF.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

## Enter book name and get suggestions of 4 books

```python
user_input = input("Enter the title of a book: ")
recommendation = recommend(user_input)
recommendation
```
